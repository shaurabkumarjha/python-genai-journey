## 1.Syntax, Indentation & PEP8

Python uses **indentation** (4 spaces) to define code blocks - no `{}` like C/Java.  
**PEP8** is Python's official style guide. Following it is expected in every MNC codebase.

**Key PEP8 Rules:**
- 4 spaces per indent level (never tabs)
- Variable names: `snake_case`
- Constants: `UPPER_CASE`
- Class names: `PascalCase`
- Max line length: 79 characters
- 2 blank lines between top-level functions/classes

````python
# PEP8 Compliant Code Example

MAX_RETRIES = 3          # Constant: UPPER_CASE
user_name = "Alice"      # Variable: snake_case
is_active = True         # Boolean variable


def greet_user(name):    # Function: snake_case, 2 blank lines before
    """Return a greeting message for the given user name."""
    return f"Hello, {name}! Welcome to Python."


class UserProfile:        # Class: PascalCase
    pass


# Calling the function
message = greet_user(user_name)
print(message)
````

## 2. Variables & Data Types

Python is **dynamically typed** - you don't declare types explicitly.

| Type | Example | `type()` Output |
|------|---------|----------------|
| `int` | `42` | `<class 'int'>` |
| `float` | `3.14` | `<class 'float'>` |
| `str` | `"hello"` | `<class 'str'>` |
| `bool` | `True` / `False` | `<class 'bool'>` |
| `list` | `[1, 2, 3]` | `<class 'list'>` |
| `tuple` | `(1, 2)` | `<class 'tuple'>` |
| `dict` | `{"k": "v"}` | `<class 'dict'>` |
| `set` | `{1, 2, 3}` | `<class 'set'>` |
| `NoneType` | `None` | `<class 'NoneType'>` |

````python
# ── Integer ──────────────────────────────────────────────────────────────────
age = 25
print(f"age        = {age}  | type: {type(age)}")

# ── Float ────────────────────────────────────────────────────────────────────
price = 99.99
print(f"price      = {price}  | type: {type(price)}")

# ── String ───────────────────────────────────────────────────────────────────
city = "Kathmandu"
print(f"city       = {city}  | type: {type(city)}")

# ── Boolean ──────────────────────────────────────────────────────────────────
is_employed = False
print(f"is_employed= {is_employed}  | type: {type(is_employed)}")

# ── List (mutable, ordered) ──────────────────────────────────────────────────
skills = ["Python", "SQL", "Docker"]
print(f"skills     = {skills}  | type: {type(skills)}")

# ── Tuple (immutable, ordered) ───────────────────────────────────────────────
coordinates = (27.7172, 85.3240)   # Kathmandu lat/long
print(f"coordinates= {coordinates}  | type: {type(coordinates)}")

# ── Dictionary (key-value pairs) ─────────────────────────────────────────────
employee = {"name": "Rahul", "role": "ML Engineer", "yoe": 3}
print(f"employee   = {employee}  | type: {type(employee)}")

# ── Set (unique, unordered) ───────────────────────────────────────────────────
tags = {"AI", "ML", "AI", "Python"}   # Duplicate 'AI' is removed
print(f"tags       = {tags}  | type: {type(tags)}")

# ── None ─────────────────────────────────────────────────────────────────────
result = None
print(f"result     = {result}  | type: {type(result)}")
````

## 3. Type Casting

Converting one data type to another. Essential when reading user input (always `str`) or working with APIs.

| Function | Converts To |
|----------|------------|
| `int()` | Integer |
| `float()` | Float |
| `str()` | String |
| `bool()` | Boolean |
| `list()` | List |
| `tuple()` | Tuple |
| `set()` | Set |

````python
# ── Implicit Casting (Python does it automatically) ───────────────────────────
x = 5        # int
y = 2.0      # float
result = x + y   # Python auto-promotes int → float
print(f"5 + 2.0 = {result}  | type: {type(result)}")  # float

print()

# ── Explicit Casting ──────────────────────────────────────────────────────────
salary_str = "85000"              # Comes from a form / API
salary_int = int(salary_str)      # str → int
salary_float = float(salary_str)  # str → float

print(f"String  : {salary_str!r}  | type: {type(salary_str)}")
print(f"Integer : {salary_int}    | type: {type(salary_int)}")
print(f"Float   : {salary_float}  | type: {type(salary_float)}")

print()

# ── bool() casting rules ──────────────────────────────────────────────────────
# Falsy: 0, 0.0, "", [], {}, set(), None
# Truthy: everything else
print(f"bool(0)    = {bool(0)}")       # False
print(f"bool(1)    = {bool(1)}")       # True
print(f"bool('')   = {bool('')}")      # False
print(f"bool('Hi') = {bool('Hi')}")   # True
print(f"bool([])   = {bool([])}")     # False

print()

# ── Safe casting with try/except ─────────────────────────────────────────────
user_input = "abc"   # Invalid number string
try:
    number = int(user_input)
except ValueError as e:
    print(f"Cannot convert '{user_input}' to int → {e}")
````

## 4. Input / Output

- `input()` — reads user input. **Always returns a string.**

## 5. Comments & Docstrings

- **Single-line comment:** `# this is a comment`
- **Inline comment:** `x = 5  # increment counter`
- **Docstring:** Triple-quoted string inside a function/class. Used by `help()`, IDEs, and auto-documentation tools like Sphinx.

MNC code reviews often check for **meaningful docstrings** on all public functions.

````python
# ── Single-line comment ───────────────────────────────────────────────────────
total = 0  # Running total of scores


def calculate_bmi(weight_kg: float, height_m: float) -> float:
    """
    Calculate Body Mass Index (BMI) from weight and height.

    Args:
        weight_kg (float): Weight in kilograms.
        height_m  (float): Height in metres.

    Returns:
        float: BMI value rounded to 2 decimal places.

    Raises:
        ValueError: If height_m is zero or negative.

    Example:
        >>> calculate_bmi(70, 1.75)
        22.86
    """
    if height_m <= 0:
        raise ValueError("Height must be a positive number.")
    return round(weight_kg / (height_m ** 2), 2)


bmi = calculate_bmi(70, 1.75)
print(f"BMI: {bmi}")

# Access the docstring programmatically
print("\n--- Docstring ---")
print(calculate_bmi.__doc__)
````

## 6. Conditional Statements — if / elif / else

Python evaluates conditions top-to-bottom and executes the **first matching block**.

```
if condition:
    ...
elif another_condition:
    ...
else:
    ...
```

**Ternary (one-liner):** `value = x if condition else y`

````python
# ── Grade Calculator ──────────────────────────────────────────────────────────
def get_grade(marks: int) -> str:
    """Return letter grade for given marks (0–100)."""
    if marks >= 90:
        return "A+"
    elif marks >= 80:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 60:
        return "C"
    elif marks >= 50:
        return "D"
    else:
        return "F"


test_scores = [95, 82, 73, 61, 55, 40]
for score in test_scores:
    print(f"Score: {score:>3}  →  Grade: {get_grade(score)}")

print()

# ── Ternary (Conditional Expression) ─────────────────────────────────────────
age = 20
status = "Adult" if age >= 18 else "Minor"
print(f"Age {age} → {status}")

# ── Logical operators: and, or, not ──────────────────────────────────────────
salary = 120000
experience = 4

if salary > 100000 and experience >= 3:
    print("Eligible for Senior role")
else:
    print("Not eligible yet")

# ── Membership & Identity operators ──────────────────────────────────────────
lang = "Python"
supported_langs = ["Python", "Go", "Rust"]

print(f"'Python' in list: {'Python' in supported_langs}")
print(f"lang is None    : {lang is None}")
````

## 7. Loops — for, while, break, continue, else

| Loop | Use Case |
|------|----------|
| `for` | Iterate over a known sequence |
| `while` | Repeat until a condition is False |
| `break` | Exit the loop immediately |
| `continue` | Skip current iteration |
| `else` on loop | Runs if loop completed without `break` |

````python
# ── for loop with range() ─────────────────────────────────────────────────────
print("Square numbers (1–5):")
for i in range(1, 6):
    print(f"  {i}² = {i**2}")

print()

# ── for loop over a list ──────────────────────────────────────────────────────
tech_stack = ["Python", "FastAPI", "Docker", "PostgreSQL", "LangChain"]
print("Tech Stack:")
for i, tech in enumerate(tech_stack, start=1):
    print(f"  {i}. {tech}")

print()

# ── while loop ────────────────────────────────────────────────────────────────
attempts = 0
max_attempts = 3
print("Login attempts:")
while attempts < max_attempts:
    attempts += 1
    print(f"  Attempt {attempts}/{max_attempts}")
print("Max attempts reached.")

print()

# ── break & continue ──────────────────────────────────────────────────────────
numbers = [3, 7, 2, 9, -1, 5, 8]
print("Processing until negative number:")
for num in numbers:
    if num < 0:
        print(f"  Found negative ({num}), stopping.")
        break
    if num % 2 == 0:
        continue   # Skip even numbers
    print(f"  Odd number: {num}")

print()

# ── Loop with else ────────────────────────────────────────────────────────────
target = 15
for n in [2, 4, 8, 16]:
    if n == target:
        print(f"Found {target}!")
        break
else:
    print(f"{target} not found in the list.")
````

## 8. Functions & Lambda

**Functions** are first-class objects in Python — they can be passed as arguments, returned, stored in variables.

Key concepts:
- Positional, keyword, default, `*args`, `**kwargs`
- Return values (can return multiple via tuple)
- Type hints (PEP484) — used in all production code
- `lambda` — anonymous one-liner function

````python
# ── Basic function with type hints ────────────────────────────────────────────
def add(a: int, b: int) -> int:
    """Return the sum of two integers."""
    return a + b

print(f"add(3, 4) = {add(3, 4)}")


# ── Default parameters ────────────────────────────────────────────────────────
def greet(name: str, role: str = "Developer") -> str:
    """Greet a user with their name and optional role."""
    return f"Hi {name}! Role: {role}"

print(greet("Sneha"))
print(greet("Ravi", role="ML Engineer"))


# ── Multiple return values (returns a tuple) ──────────────────────────────────
def min_max(numbers: list) -> tuple:
    """Return the minimum and maximum values of a list."""
    return min(numbers), max(numbers)

lo, hi = min_max([4, 1, 9, 3, 7])
print(f"Min: {lo} | Max: {hi}")


# ── *args — variable positional arguments ─────────────────────────────────────
def total_cost(*prices: float) -> float:
    """Return the sum of any number of price values."""
    return sum(prices)

print(f"Total: ₹{total_cost(299, 499, 149, 99):.2f}")


# ── **kwargs — variable keyword arguments ────────────────────────────────────
def create_profile(**info) -> dict:
    """Build a user profile dict from keyword arguments."""
    return info

profile = create_profile(name="Kavya", city="Pune", skill="Python")
print(f"Profile: {profile}")


# ── Lambda function ───────────────────────────────────────────────────────────
# Syntax: lambda arguments: expression

square = lambda x: x ** 2
print(f"square(9) = {square(9)}")

# Common use: sorting a list of dicts by a key
employees = [
    {"name": "Alice", "salary": 90000},
    {"name": "Bob",   "salary": 75000},
    {"name": "Carol", "salary": 110000},
]
sorted_employees = sorted(employees, key=lambda e: e["salary"], reverse=True)
print("\nEmployees by salary (descending):")
for emp in sorted_employees:
    print(f"  {emp['name']:<8} ₹{emp['salary']:,}")
````

## 9. List Comprehensions

A concise, Pythonic way to create lists. Replaces multi-line for-loops.

```python
# Syntax
new_list = [expression for item in iterable if condition]
```

> 💡 Comprehensions are faster than equivalent `for` + `append` loops (CPython optimized).

````python
# ── Basic list comprehension ──────────────────────────────────────────────────
squares = [x ** 2 for x in range(1, 11)]
print(f"Squares  : {squares}")

# ── With filter condition ─────────────────────────────────────────────────────
even_squares = [x ** 2 for x in range(1, 11) if x % 2 == 0]
print(f"Even²    : {even_squares}")

# ── String transformation ─────────────────────────────────────────────────────
raw_names = [" alice ", "BOB", "  Carol  "]
clean_names = [name.strip().title() for name in raw_names]
print(f"Cleaned  : {clean_names}")

# ── Nested list comprehension (flatten a 2D list) ─────────────────────────────
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [num for row in matrix for num in row]
print(f"Flattened: {flat}")

# ── if-else inside comprehension ─────────────────────────────────────────────
scores = [88, 42, 95, 60, 37, 75]
result = ["PASS" if s >= 50 else "FAIL" for s in scores]
print(f"Results  : {result}")

# ── Comparison: for-loop vs comprehension ────────────────────────────────────
# Traditional way
traditional = []
for x in range(10):
    if x % 3 == 0:
        traditional.append(x * 2)

# Pythonic way
pythonic = [x * 2 for x in range(10) if x % 3 == 0]

print(f"\nTraditional : {traditional}")
print(f"Pythonic    : {pythonic}")
print(f"Equal?      : {traditional == pythonic}")
````

## 10. Dict & Set Comprehensions

Same concept as list comprehensions — just different bracket syntax.

```python
# Dict comprehension
{key_expr: value_expr for item in iterable if condition}

# Set comprehension  
{expression for item in iterable if condition}
```

````python
# ── Dict Comprehension ────────────────────────────────────────────────────────

# Build a word → length mapping
words = ["Python", "AI", "Machine", "Learning", "LLM"]
word_lengths = {word: len(word) for word in words}
print(f"Word lengths : {word_lengths}")

# Invert a dictionary (swap keys and values)
original = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in original.items()}
print(f"Original     : {original}")
print(f"Inverted     : {inverted}")

# Filter a dict — keep only high earners
salaries = {"Alice": 120000, "Bob": 85000, "Carol": 150000, "Dave": 72000}
high_earners = {name: sal for name, sal in salaries.items() if sal > 90000}
print(f"High earners : {high_earners}")

print()

# ── Set Comprehension ─────────────────────────────────────────────────────────

# Unique first letters from a word list
words_2 = ["apple", "banana", "avocado", "cherry", "blueberry", "apricot"]
first_letters = {word[0] for word in words_2}
print(f"First letters (unique) : {sorted(first_letters)}")

# Remove duplicates while transforming
nums = [1, 2, 2, 3, 4, 4, 5]
unique_squares = {x ** 2 for x in nums}
print(f"Unique squares         : {sorted(unique_squares)}")
````

## Practice Exercises

1. Write a function `is_palindrome(s: str) -> bool` that returns `True` if string `s` is a palindrome (ignore case, ignore spaces).
2. Using list comprehension, generate all **Pythagorean triplets** `(a, b, c)` where `a, b, c <= 20`.
3. Given `text = "the quick brown fox jumps over the lazy dog"`, use a dict comprehension to count the frequency of each word.
4. Write a function `flatten_and_deduplicate(lists)` that takes a list of lists, flattens it, removes duplicates, and returns a sorted list.

````python
# ── Exercise Solutions ────────────────────────────────────────────────────────

# 1. Palindrome check
def is_palindrome(s: str) -> bool:
    """Return True if s is a palindrome (case-insensitive, space-ignored)."""
    cleaned = s.replace(" ", "").lower()
    return cleaned == cleaned[::-1]

test_words = ["racecar", "hello", "A man a plan a canal Panama", "Python"]
for w in test_words:
    print(f"is_palindrome({w!r}) = {is_palindrome(w)}")

print()

# 2. Pythagorean triplets
triplets = [
    (a, b, c)
    for a in range(1, 21)
    for b in range(a, 21)
    for c in range(b, 21)
    if a**2 + b**2 == c**2
]
print(f"Pythagorean triplets (≤20): {triplets}")

print()

# 3. Word frequency counter
text = "the quick brown fox jumps over the lazy dog"
word_freq = {word: text.split().count(word) for word in set(text.split())}
# Sort by frequency descending
sorted_freq = dict(sorted(word_freq.items(), key=lambda x: x[1], reverse=True))
print("Word frequencies:", sorted_freq)

print()

# 4. Flatten and deduplicate
def flatten_and_deduplicate(lists: list) -> list:
    """Flatten a list of lists, remove duplicates, return sorted."""
    return sorted({item for sublist in lists for item in sublist})

sample = [[3, 1, 2], [2, 4, 5], [1, 6, 3, 7]]
print(f"Input : {sample}")
print(f"Output: {flatten_and_deduplicate(sample)}")
````
