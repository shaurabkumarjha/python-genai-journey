# PART 1 - CLASSES & `__init__`

## Theory

A **class** is a blueprint. An **object** (instance) is the thing built from that blueprint.

| Term | Meaning |
|------|---------|
| Class | Blueprint / template |
| Instance | A specific object created from the class |
| `self` | Reference to the current instance |
| `__init__` | Constructor - called automatically when instance is created |
| Instance variable | Unique to each instance - defined with `self.x` |
| Class variable | Shared by ALL instances - defined directly in class body |

```python
# ═══════════════════════════════════════════════════════════════════════════════
# 1.1  CLASS BASICS - instance vs class variables
# ═══════════════════════════════════════════════════════════════════════════════

class Employee:
    """
    Represents a company employee.
    Demonstrates instance variables vs class variables.
    """

    # ── Class variables - shared by ALL instances ─────────────────────────────
    company_name  = "TechCorp"           # Same for everyone
    employee_count = 0                   # Tracks total employees
    CURRENCY      = "INR"                # Constant (by convention UPPER_CASE)

    # ── Constructor ───────────────────────────────────────────────────────────
    def __init__(self, name: str, role: str, salary: float):
        """
        Called automatically when Employee(...) is invoked.
        self = the new instance being created.
        """
        # Instance variables - unique to each employee
        self.name    = name
        self.role    = role
        self.salary  = salary
        self.is_active = True

        # Update class variable each time an employee is created
        Employee.employee_count += 1

    # ── Instance method ───────────────────────────────────────────────────────
    def give_raise(self, amount: float) -> None:
        """Increase salary by given amount."""
        self.salary += amount
        print(f"  {self.name} got a raise! New salary: {self.salary:,.0f} {Employee.CURRENCY}")

    def get_info(self) -> str:
        """Return formatted employee summary."""
        status = "Active" if self.is_active else "Inactive"
        return f"{self.name} | {self.role} | {self.salary:,.0f} {Employee.CURRENCY} | {status}"


# Creating instances
e1 = Employee("Alice", "ML Engineer",    120000)
e2 = Employee("Bob",   "Data Analyst",    85000)
e3 = Employee("Carol", "DevOps Engineer", 100000)

print(f"Total employees: {Employee.employee_count}")   # Class variable
print(f"Company: {e1.company_name}")                    # Accessible on instance too
print()
print(e1.get_info())
print(e2.get_info())

e1.give_raise(20000)

# Instance var vs class var - tricky behaviour
e1.company_name = "StartupXYZ"        # Creates INSTANCE var, does NOT change class var
print(f"\ne1.company_name = {e1.company_name}")
print(f"e2.company_name = {e2.company_name}")   # Still TechCorp
print(f"Employee.company_name = {Employee.company_name}")  # Still TechCorp
```

# PART 2 - DUNDER (MAGIC) METHODS

## Theory

Dunder = **D**ouble **under**score. Also called "magic methods" or "special methods".
They let your objects work with Python's built-in operations like `+`, `len()`, `print()`, `==`, `[]`.

| Dunder | Triggered by | Purpose |
|--------|-------------|--------|
| `__init__` | `ClassName()` | Constructor |
| `__repr__` | `repr(obj)`, debugger | Official string representation |
| `__str__` | `str(obj)`, `print(obj)` | Human-readable string |
| `__len__` | `len(obj)` | Length |
| `__getitem__` | `obj[key]` | Index/key access |
| `__setitem__` | `obj[key] = val` | Index assignment |
| `__delitem__` | `del obj[key]` | Delete by index |
| `__contains__` | `x in obj` | Membership test |
| `__iter__` | `for x in obj` | Iterator |
| `__next__` | `next(obj)` | Next value |
| `__add__` | `obj + other` | Addition |
| `__mul__` | `obj * n` | Multiplication |
| `__eq__` | `obj == other` | Equality |
| `__lt__` | `obj < other` | Less than |
| `__le__`, `__gt__`, `__ge__`, `__ne__` | `<=`, `>`, `>=`, `!=` | Comparisons |
| `__hash__` | `hash(obj)`, dict key | Hash value |
| `__bool__` | `bool(obj)`, `if obj:` | Truthiness |
| `__call__` | `obj()` | Make object callable |
| `__enter__`/`__exit__` | `with obj:` | Context manager |
| `__del__` | `del obj` | Destructor |

```python
# ═══════════════════════════════════════════════════════════════════════════════
# 2.1  CORE DUNDER METHODS
# ═══════════════════════════════════════════════════════════════════════════════

class Vector:
    """
    2D mathematical vector.
    Demonstrates most common dunder methods.
    """

    def __init__(self, x: float, y: float):
        self.x = x
        self.y = y

    # ── String representations ────────────────────────────────────────────────
    def __repr__(self) -> str:
        """Used in REPL, repr(), debugging. Should be unambiguous."""
        return f"Vector({self.x}, {self.y})"

    def __str__(self) -> str:
        """Used by print(). Human-readable. Falls back to __repr__ if missing."""
        return f"({self.x}, {self.y})"

    # ── Arithmetic operators ──────────────────────────────────────────────────
    def __add__(self, other: "Vector") -> "Vector":
        """v1 + v2"""
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other: "Vector") -> "Vector":
        """v1 - v2"""
        return Vector(self.x - other.x, self.y - other.y)

    def __mul__(self, scalar: float) -> "Vector":
        """v * 3 (scalar multiplication)"""
        return Vector(self.x * scalar, self.y * scalar)

    def __rmul__(self, scalar: float) -> "Vector":
        """3 * v (reflected - when left side doesn't support the op)"""
        return self.__mul__(scalar)

    def __neg__(self) -> "Vector":
        """-v (unary negation)"""
        return Vector(-self.x, -self.y)

    def __abs__(self) -> float:
        """abs(v) → magnitude"""
        return (self.x**2 + self.y**2) ** 0.5

    # ── Comparison operators ──────────────────────────────────────────────────
    def __eq__(self, other: object) -> bool:
        """v1 == v2"""
        if not isinstance(other, Vector):
            return NotImplemented
        return self.x == other.x and self.y == other.y

    def __lt__(self, other: "Vector") -> bool:
        """v1 < v2 (by magnitude)"""
        return abs(self) < abs(other)

    def __le__(self, other: "Vector") -> bool:
        return abs(self) <= abs(other)

    # ── Boolean, hash, len ────────────────────────────────────────────────────
    def __bool__(self) -> bool:
        """bool(v) - False if zero vector"""
        return self.x != 0 or self.y != 0

    def __hash__(self) -> int:
        """hash(v) - required to use as dict key or set member"""
        return hash((self.x, self.y))


v1 = Vector(3, 4)
v2 = Vector(1, 2)
v3 = Vector(0, 0)

print(f"repr(v1)    = {repr(v1)}")
print(f"print(v1)   = {v1}")
print(f"v1 + v2     = {v1 + v2}")
print(f"v1 - v2     = {v1 - v2}")
print(f"v1 * 3      = {v1 * 3}")
print(f"2 * v1      = {2 * v1}")
print(f"-v1         = {-v1}")
print(f"abs(v1)     = {abs(v1):.2f}")
print(f"v1 == v2    = {v1 == v2}")
print(f"v2 < v1     = {v2 < v1}")
print(f"bool(v1)    = {bool(v1)}")
print(f"bool(v3)    = {bool(v3)}")

# Usable in set/dict because __hash__ defined
vectors_set = {v1, v2, v1}     # v1 appears twice but set deduplicates
print(f"Set of vectors: {vectors_set}")

# Sort list of vectors (uses __lt__)
vecs = [Vector(3, 4), Vector(1, 0), Vector(2, 2)]
sorted_vecs = sorted(vecs)
print(f"\nSorted by magnitude: {sorted_vecs}")
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# 2.2  CONTAINER DUNDERS - make objects behave like lists/dicts
# ═══════════════════════════════════════════════════════════════════════════════

class Playlist:
    """
    Music playlist that behaves like a sequence.
    Demonstrates container dunder methods.
    """

    def __init__(self, name: str):
        self.name   = name
        self._songs = []        # "private" - convention only

    def add(self, song: str) -> None:
        self._songs.append(song)

    def __len__(self) -> int:
        """len(playlist)"""
        return len(self._songs)

    def __getitem__(self, index):
        """playlist[0]  or  playlist[1:3]"""
        return self._songs[index]

    def __setitem__(self, index, song: str):
        """playlist[0] = 'new song'"""
        self._songs[index] = song

    def __delitem__(self, index):
        """del playlist[0]"""
        del self._songs[index]

    def __contains__(self, song: str) -> bool:
        """'song' in playlist"""
        return song in self._songs

    def __iter__(self):
        """for song in playlist:"""
        return iter(self._songs)

    def __reversed__(self):
        """reversed(playlist)"""
        return reversed(self._songs)

    def __repr__(self) -> str:
        return f"Playlist({self.name!r}, {self._songs})"

    def __str__(self) -> str:
        songs_str = "\n".join(f"  {i+1}. {s}" for i, s in enumerate(self._songs))
        return f"🎵 {self.name}\n{songs_str}"


pl = Playlist("My Fav")
for song in ["Shape of You", "Blinding Lights", "Levitating", "As It Was"]:
    pl.add(song)

print(pl)                                   # __str__
print(f"\nlen     = {len(pl)}")             # __len__
print(f"[0]     = {pl[0]}")                 # __getitem__
print(f"[1:3]   = {pl[1:3]}")              # __getitem__ slice
print(f"'Levitating' in pl = {'Levitating' in pl}")  # __contains__

pl[0] = "Stay"                              # __setitem__
print(f"After pl[0]='Stay': {pl[0]}")

print("\nIteration:")
for song in pl:                             # __iter__
    print(f"  ♪ {song}")

print("\nReversed:")
for song in reversed(pl):                  # __reversed__
    print(f"  ♪ {song}")
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# 2.3  __call__, __enter__/__exit__, __del__
# ═══════════════════════════════════════════════════════════════════════════════

# ── __call__ - make instance callable like a function ─────────────────────────
class Multiplier:
    """Creates a callable multiplier with a fixed factor."""

    def __init__(self, factor: float):
        self.factor = factor

    def __call__(self, value: float) -> float:
        """Called when instance is used as a function: obj(arg)."""
        return value * self.factor

    def __repr__(self):
        return f"Multiplier(factor={self.factor})"


double    = Multiplier(2)
triple    = Multiplier(3)
to_dozen  = Multiplier(12)

print(f"double(5)    = {double(5)}")
print(f"triple(7)    = {triple(7)}")
print(f"to_dozen(4)  = {to_dozen(4)}")
print(f"callable?    = {callable(double)}")

# ── __enter__ / __exit__ - context manager ────────────────────────────────────
class DatabaseConnection:
    """Simulates a DB connection as a context manager."""

    def __init__(self, db_name: str):
        self.db_name   = db_name
        self.connected = False

    def __enter__(self):
        """Called at start of `with` block. Returns the resource."""
        self.connected = True
        print(f"  [DB] Connected to '{self.db_name}'")
        return self           # this becomes `as` variable

    def __exit__(self, exc_type, exc_val, exc_tb):
        """
        Called at end of `with` block (even if exception occurred).
        Return True to suppress exceptions, False to propagate.
        """
        self.connected = False
        if exc_type:
            print(f"  [DB] Error occurred: {exc_val}")
        print(f"  [DB] Connection to '{self.db_name}' closed.")
        return False   # Don't suppress exceptions

    def query(self, sql: str):
        if not self.connected:
            raise RuntimeError("Not connected!")
        print(f"  [DB] Executing: {sql}")
        return [{"id": 1, "name": "Alice"}, {"id": 2, "name": "Bob"}]


print()
with DatabaseConnection("users_db") as db:
    results = db.query("SELECT * FROM users")
    print(f"  [DB] Got {len(results)} rows")

print(f"  Connected after block? {db.connected}")
```

# PART 3 - PROPERTIES - `@property`

## Theory

`@property` lets you define getter/setter/deleter methods that are **accessed like attributes**.  
Benefit: You can add validation logic without changing the public interface.

```python
# ═══════════════════════════════════════════════════════════════════════════════
# 3.1  @property, setter, deleter
# ═══════════════════════════════════════════════════════════════════════════════

class BankAccount:
    """
    Bank account with property-based validation.
    Balance can't go negative. Owner name can't be empty.
    """

    def __init__(self, owner: str, initial_balance: float = 0):
        self._owner   = None    # Use private var, set via property
        self._balance = 0.0
        self.owner    = owner           # Triggers the setter
        self.deposit(initial_balance)

    # ── owner property ────────────────────────────────────────────────────────
    @property
    def owner(self) -> str:
        """Getter - called when you read account.owner."""
        return self._owner

    @owner.setter
    def owner(self, name: str) -> None:
        """Setter - called when you do account.owner = 'Alice'."""
        if not name or not name.strip():
            raise ValueError("Owner name cannot be empty.")
        self._owner = name.strip().title()

    # ── balance property ──────────────────────────────────────────────────────
    @property
    def balance(self) -> float:
        """Read-only property - no setter defined, so balance can't be set directly."""
        return self._balance

    # ── computed property (no backing variable) ───────────────────────────────
    @property
    def balance_usd(self) -> float:
        """Computed from balance at read time. No storage needed."""
        return round(self._balance / 83.5, 2)   # Approx INR → USD

    # ── Regular methods ───────────────────────────────────────────────────────
    def deposit(self, amount: float) -> None:
        if amount < 0:
            raise ValueError("Deposit amount must be positive.")
        self._balance += amount

    def withdraw(self, amount: float) -> None:
        if amount > self._balance:
            raise ValueError(f"Insufficient funds. Balance: {self._balance:.2f}")
        self._balance -= amount

    def __repr__(self):
        return f"BankAccount(owner={self.owner!r}, balance={self.balance:.2f})"


acc = BankAccount("alice sharma", 50000)
print(f"Account : {acc}")
print(f"Balance : ₹{acc.balance:,.2f}")
print(f"In USD  : ${acc.balance_usd}")

acc.deposit(10000)
acc.withdraw(5000)
print(f"After tx: ₹{acc.balance:,.2f}")

# Property setter with validation
acc.owner = "  rahul verma  "   # Extra whitespace - gets cleaned
print(f"Owner: {acc.owner}")

# Read-only property
try:
    acc.balance = 999999          # No setter → AttributeError
except AttributeError as e:
    print(f"Can't set balance directly: {e}")

# Validation in setter
try:
    acc.owner = ""                # Empty name
except ValueError as e:
    print(f"ValueError: {e}")
```

# PART 4 - CLASS METHODS & STATIC METHODS

| Decorator | First param | Access to | Use case |
|-----------|-------------|-----------|----------|
| (none) | `self` | instance + class | Regular operations on instance |
| `@classmethod` | `cls` | class only | Alternative constructors, factory methods |
| `@staticmethod` | (none) | neither | Utility functions related to the class |

```python
# ═══════════════════════════════════════════════════════════════════════════════
# 4.1  @classmethod and @staticmethod
# ═══════════════════════════════════════════════════════════════════════════════

class Temperature:
    """Temperature in Celsius with class/static method examples."""

    ABSOLUTE_ZERO_C = -273.15

    def __init__(self, celsius: float):
        self.celsius = celsius

    # ── Instance method ───────────────────────────────────────────────────────
    def to_fahrenheit(self) -> float:
        """Needs self - uses instance data."""
        return self.celsius * 9/5 + 32

    def to_kelvin(self) -> float:
        return self.celsius - Temperature.ABSOLUTE_ZERO_C

    # ── Class methods - alternative constructors ──────────────────────────────
    @classmethod
    def from_fahrenheit(cls, f: float) -> "Temperature":
        """Factory: create Temperature from Fahrenheit."""
        return cls((f - 32) * 5/9)

    @classmethod
    def from_kelvin(cls, k: float) -> "Temperature":
        """Factory: create Temperature from Kelvin."""
        return cls(k + Temperature.ABSOLUTE_ZERO_C)

    @classmethod
    def absolute_zero(cls) -> "Temperature":
        """Returns the absolute zero temperature."""
        return cls(Temperature.ABSOLUTE_ZERO_C)

    # ── Static methods - utility functions ────────────────────────────────────
    @staticmethod
    def is_valid_celsius(c: float) -> bool:
        """Validation - doesn't need self or cls."""
        return c >= Temperature.ABSOLUTE_ZERO_C

    @staticmethod
    def celsius_to_fahrenheit(c: float) -> float:
        """Conversion utility - doesn't need an instance."""
        return c * 9/5 + 32

    def __repr__(self):
        return f"Temperature({self.celsius:.2f}°C)"


# Create via regular constructor
t1 = Temperature(100)
print(f"t1 = {t1}")
print(f"Fahrenheit: {t1.to_fahrenheit():.1f}°F")
print(f"Kelvin    : {t1.to_kelvin():.2f} K")

# Create via class methods (alternative constructors)
t2 = Temperature.from_fahrenheit(212)
t3 = Temperature.from_kelvin(373.15)
az = Temperature.absolute_zero()

print(f"\nFrom 212°F       : {t2}")
print(f"From 373.15 K    : {t3}")
print(f"Absolute zero    : {az}")

# Static methods - call on class or instance
print(f"\nValid -300°C?    : {Temperature.is_valid_celsius(-300)}")
print(f"Valid -270°C?    : {Temperature.is_valid_celsius(-270)}")
print(f"0°C in °F        : {Temperature.celsius_to_fahrenheit(0)}")
```

# PART 5 - INHERITANCE

## Theory

Inheritance = a child class **inherits** attributes and methods from a parent class.

| Type | Syntax | Example |
|------|--------|---------|
| Single | `class Dog(Animal)` | Dog extends Animal |
| Multi-level | `class Puppy(Dog)` | Puppy → Dog → Animal |
| Multiple | `class C(A, B)` | C inherits from both A and B |

`super()` calls the parent class method - essential in `__init__` chains.

```python
# ═══════════════════════════════════════════════════════════════════════════════
# 5.1  SINGLE INHERITANCE + super()
# ═══════════════════════════════════════════════════════════════════════════════

class Animal:
    """
    Base class - defines the common interface all animals share.
    """

    def __init__(self, name: str, species: str, age: int):
        self.name    = name
        self.species = species
        self.age     = age
        self._health = 100    # protected attribute

    def speak(self) -> str:
        """To be overridden by subclasses."""
        return f"{self.name} makes a sound."

    def eat(self, food: str) -> str:
        return f"{self.name} eats {food}."

    def info(self) -> str:
        return f"{self.name} ({self.species}, {self.age} yrs, health={self._health}%)"

    def __repr__(self):
        return f"{self.__class__.__name__}({self.name!r})"

    def __str__(self):
        return self.info()


class Dog(Animal):
    """Dog - extends Animal with dog-specific behaviour."""

    def __init__(self, name: str, age: int, breed: str):
        # super() calls Animal.__init__ - always call it first
        super().__init__(name, species="Canis lupus familiaris", age=age)
        self.breed  = breed
        self.tricks = []

    def speak(self) -> str:
        """Override parent's speak()."""
        return f"{self.name} says: Woof! Woof!"

    def learn_trick(self, trick: str) -> str:
        """Dog-specific method."""
        self.tricks.append(trick)
        return f"{self.name} learned: {trick}!"

    def perform(self) -> str:
        if not self.tricks:
            return f"{self.name} doesn't know any tricks yet."
        return f"{self.name} performs: {', '.join(self.tricks)}"

    def info(self) -> str:
        """Extend parent's info() with breed."""
        base = super().info()        # Get parent's string
        return f"{base}, breed={self.breed}"


class Cat(Animal):
    """Cat - extends Animal with cat-specific behaviour."""

    def __init__(self, name: str, age: int, indoor: bool = True):
        super().__init__(name, species="Felis catus", age=age)
        self.indoor     = indoor
        self.lives_left = 9

    def speak(self) -> str:
        return f"{self.name} says: Meow"

    def purr(self) -> str:
        return f"{self.name} purrs contentedly... purrrrr"

    def info(self) -> str:
        base = super().info()
        lifestyle = "Indoor" if self.indoor else "Outdoor"
        return f"{base}, {lifestyle}, lives={self.lives_left}"


# Create instances
dog = Dog("Bruno", 3, "Labrador")
cat = Cat("Whiskers", 5, indoor=True)

print("── Dog ──")
print(dog)
print(dog.speak())
print(dog.eat("chicken"))
print(dog.learn_trick("sit"))
print(dog.learn_trick("paw"))
print(dog.perform())

print("\n── Cat ──")
print(cat)
print(cat.speak())
print(cat.purr())

# isinstance - checks inheritance chain
print(f"\ndog is Dog?    {isinstance(dog, Dog)}")
print(f"dog is Animal? {isinstance(dog, Animal)}")   # True! inherits
print(f"dog is Cat?    {isinstance(dog, Cat)}")

# issubclass
print(f"Dog is subclass of Animal? {issubclass(Dog, Animal)}")
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# 5.2  MULTI-LEVEL INHERITANCE
# ═══════════════════════════════════════════════════════════════════════════════

class GuideDog(Dog):
    """Guide dog - extends Dog, which already extends Animal."""

    def __init__(self, name: str, age: int, breed: str, owner_name: str):
        super().__init__(name, age, breed)     # Calls Dog.__init__
        self.owner_name   = owner_name
        self.is_certified = False

    def certify(self) -> str:
        self.is_certified = True
        return f"{self.name} is now a certified guide dog for {self.owner_name}!"

    def guide(self) -> str:
        if not self.is_certified:
            return f"{self.name} is not yet certified."
        return f"{self.name} guides {self.owner_name} safely."


gd = GuideDog("Rex", 4, "Golden Retriever", "Ravi")
print(f"MRO: {[c.__name__ for c in GuideDog.__mro__]}")
print(gd.speak())          # From Dog
print(gd.eat("kibble"))    # From Animal
print(gd.learn_trick("stop at curb"))  # From Dog
print(gd.certify())
print(gd.guide())
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# 5.3  MULTIPLE INHERITANCE & MRO (Method Resolution Order)
# ═══════════════════════════════════════════════════════════════════════════════

class Flyable:
    """Mixin - adds flying capability."""

    def fly(self) -> str:
        return f"{self.name} is flying! ✈️"

    def describe_movement(self) -> str:
        return "Moves through the air."


class Swimmable:
    """Mixin - adds swimming capability."""

    def swim(self) -> str:
        return f"{self.name} is swimming! "

    def describe_movement(self) -> str:
        return "Moves through water."


class Duck(Animal, Flyable, Swimmable):
    """Duck can fly AND swim - multiple inheritance."""

    def __init__(self, name: str, age: int):
        # Only call Animal.__init__ - mixins don't have __init__
        super().__init__(name, species="Anas platyrhynchos", age=age)

    def speak(self) -> str:
        return f"{self.name} says: Quack!"


duck = Duck("Donald", 2)
print(duck.speak())
print(duck.fly())
print(duck.swim())

# MRO - which describe_movement() gets called?
# Python uses C3 linearization (left to right, depth first)
print(f"\nMRO: {[c.__name__ for c in Duck.__mro__]}")
print(f"describe_movement(): {duck.describe_movement()}")  # From Flyable (listed first)

# inspect MRO
import inspect
print(f"\nduck.fly defined in: {inspect.getmro(Duck)}")
```

# PART 6 - POLYMORPHISM

## Theory

**Polymorphism** = many forms. Same interface, different behaviour depending on type.

Two flavours in Python:
1. **Method overriding** - subclass provides its own version of a parent method.
2. **Duck typing** - "If it walks like a duck and quacks like a duck, it's a duck." Python doesn't check type, just checks if the method exists.

```python
# ═══════════════════════════════════════════════════════════════════════════════
# 6.1  POLYMORPHISM - method overriding + duck typing
# ═══════════════════════════════════════════════════════════════════════════════

# All speak() differently - same call, different result
animals = [
    Dog("Bruno", 3, "Labrador"),
    Cat("Whiskers", 5),
    Duck("Donald", 2),
    Animal("Mystery", "Unknown", 1),  # Base class
]

print("Polymorphic speak():")
for animal in animals:
    # Same code, different behaviour depending on actual type
    print(f"  {type(animal).__name__:<10}: {animal.speak()}")


# ── Duck typing - no type checking needed ─────────────────────────────────────
class Robot:
    """Not an Animal at all - but has speak() method."""
    name = "R2D2"

    def speak(self) -> str:
        return f"{self.name} says: Beep boop!"


class Human:
    def __init__(self, name):
        self.name = name

    def speak(self) -> str:
        return f"{self.name} says: Hello, world!"


def make_everyone_speak(speakers):
    """Works with ANY object that has a speak() method - duck typing."""
    for speaker in speakers:
        print(f"  {speaker.speak()}")


print("\nDuck typing - different types, same interface:")
make_everyone_speak([Dog("Max", 2, "Poodle"), Cat("Luna", 3), Robot(), Human("Alice")])

# ── Abstract base class - enforce interface ────────────────────────────────────
from abc import ABC, abstractmethod

class Shape(ABC):
    """Abstract base class - cannot be instantiated directly."""

    @abstractmethod
    def area(self) -> float:
        """Must be implemented by every subclass."""
        pass

    @abstractmethod
    def perimeter(self) -> float:
        pass

    def describe(self) -> str:
        return f"{self.__class__.__name__}: area={self.area():.2f}, perimeter={self.perimeter():.2f}"


class Circle(Shape):
    import math
    def __init__(self, radius: float):
        self.radius = radius
    def area(self)      -> float: import math; return math.pi * self.radius**2
    def perimeter(self) -> float: import math; return 2 * math.pi * self.radius


class Rectangle(Shape):
    def __init__(self, w: float, h: float):
        self.w, self.h = w, h
    def area(self)      -> float: return self.w * self.h
    def perimeter(self) -> float: return 2 * (self.w + self.h)


class Triangle(Shape):
    def __init__(self, a: float, b: float, c: float):
        self.a, self.b, self.c = a, b, c
    def area(self) -> float:
        s = (self.a + self.b + self.c) / 2
        return (s * (s-self.a) * (s-self.b) * (s-self.c)) ** 0.5
    def perimeter(self) -> float: return self.a + self.b + self.c


shapes = [Circle(5), Rectangle(4, 6), Triangle(3, 4, 5)]
print("\nPolymorphic shapes:")
for shape in shapes:
    print(f"  {shape.describe()}")

total_area = sum(s.area() for s in shapes)
print(f"  Total area: {total_area:.2f}")

# Can't instantiate abstract class
try:
    s = Shape()
except TypeError as e:
    print(f"\nCan't instantiate ABC: {e}")
```

# PART 7 - ENCAPSULATION

## Theory

Encapsulation = hiding internal details, exposing only what's necessary.

| Convention | Syntax | Meaning |
|------------|--------|---------|
| Public | `self.name` | Anyone can access |
| Protected | `self._name` | Convention: internal use / subclasses |
| Private | `self.__name` | Name mangled → `_ClassName__name` - harder to access |

Python doesn't enforce access control like Java/C++. It relies on conventions.

```python
# ═══════════════════════════════════════════════════════════════════════════════
# 7.1  PUBLIC, PROTECTED, PRIVATE
# ═══════════════════════════════════════════════════════════════════════════════

class SecureVault:
    """Demonstrates all three access levels."""

    MAX_ATTEMPTS = 3                  # Public class constant

    def __init__(self, owner: str, pin: str):
        self.owner         = owner   # Public
        self._access_log   = []      # Protected - for subclasses to use
        self.__pin         = pin     # Private - name mangled to _SecureVault__pin
        self.__attempts    = 0
        self.__is_locked   = False

    def unlock(self, pin: str) -> bool:
        """Public method - safe interface to private data."""
        if self.__is_locked:
            print("  Vault is locked out!")
            return False

        if pin == self.__pin:
            self.__attempts = 0
            self._access_log.append(f"Unlocked by {self.owner}")
            print("Vault unlocked!")
            return True
        else:
            self.__attempts += 1
            remaining = SecureVault.MAX_ATTEMPTS - self.__attempts
            print(f"Wrong PIN. {remaining} attempts left.")
            if self.__attempts >= SecureVault.MAX_ATTEMPTS:
                self.__is_locked = True
                print("Vault locked out!")
            return False

    def change_pin(self, old_pin: str, new_pin: str) -> bool:
        """Encapsulated: can only change pin if you know the old one."""
        if old_pin == self.__pin:
            self.__pin = new_pin
            print("  PIN changed successfully.")
            return True
        print("  Wrong old PIN.")
        return False

    @property
    def is_locked(self) -> bool:
        return self.__is_locked


vault = SecureVault("Alice", "1234")
vault.unlock("0000")          # Wrong
vault.unlock("0000")          # Wrong
vault.unlock("1234")          # Correct
vault.change_pin("1234", "9999")
vault.unlock("9999")          # New pin works

# Accessing private via name mangling (possible but strongly discouraged)
print(f"\nName mangled:  vault._SecureVault__pin = {vault._SecureVault__pin}")
print(f"Direct access: vault.__pin would fail:")
try:
    print(vault.__pin)
except AttributeError as e:
    print(f"  AttributeError: {e}")
```

# PART 8 - DATACLASSES

## Theory

`@dataclass` (Python 3.7+) auto-generates boilerplate: `__init__`, `__repr__`, `__eq__`.
Use when your class is mainly a data container.

| Parameter | Default | Effect |
|-----------|---------|--------|
| `repr=True` | True | Auto `__repr__` |
| `eq=True` | True | Auto `__eq__` |
| `order=False` | False | Auto `<`, `>`, `<=`, `>=` |
| `frozen=False` | False | Immutable if True |
| `slots=False` | False | Use `__slots__` (Python 3.10+) |

```python
# ═══════════════════════════════════════════════════════════════════════════════
# 8.1  DATACLASSES - modern, clean data containers
# ═══════════════════════════════════════════════════════════════════════════════

from dataclasses import dataclass, field, asdict, astuple
from typing import List


@dataclass
class Point:
    """Simple 2D point - minimal dataclass."""
    x: float
    y: float


@dataclass(order=True, frozen=True)
class Version:
    """
    Semantic version (immutable + orderable).
    frozen=True makes it immutable (like a namedtuple).
    order=True enables <, >, <=, >= comparisons.
    """
    major: int
    minor: int
    patch: int = 0

    def __str__(self):
        return f"v{self.major}.{self.minor}.{self.patch}"


@dataclass
class Student:
    """
    Student with default factory, computed field, and post-init logic.
    """
    name        : str
    roll        : int
    subjects    : List[str] = field(default_factory=list)  # Mutable default!
    scores      : dict      = field(default_factory=dict)
    _grade      : str       = field(init=False, repr=False) # Computed, not in __init__

    def __post_init__(self):
        """Called after __init__ - for validation and computed fields."""
        if not self.name.strip():
            raise ValueError("Name cannot be empty.")
        self._grade = self._compute_grade()

    def _compute_grade(self) -> str:
        if not self.scores:
            return "N/A"
        avg = sum(self.scores.values()) / len(self.scores)
        if avg >= 90: return "A+"
        elif avg >= 80: return "A"
        elif avg >= 70: return "B"
        elif avg >= 60: return "C"
        else: return "F"

    @property
    def grade(self):
        return self._grade

    def add_score(self, subject: str, score: int):
        self.scores[subject] = score
        self._grade = self._compute_grade()


# ── Usage ─────────────────────────────────────────────────────────────────────
p = Point(3.0, 4.0)
print(f"Point: {p}")
print(f"x={p.x}, y={p.y}")

v1 = Version(1, 2, 3)
v2 = Version(2, 0)
print(f"\nv1={v1}, v2={v2}")
print(f"v1 < v2: {v1 < v2}")
print(f"Sorted: {sorted([Version(1,9), Version(2,0), Version(1,0,5)])}")

try:
    v1.major = 99    # frozen=True prevents mutation
except Exception as e:
    print(f"Frozen: {e}")

s = Student("Priya", 101)
for subject, score in [("Math", 92), ("CS", 96), ("English", 88)]:
    s.add_score(subject, score)
    s.subjects.append(subject)

print(f"\n{s}")
print(f"Grade: {s.grade}")
print(f"As dict: {asdict(s)}")
```

# PART 9 - FULL CLASS HIERARCHY
### `Animal → Dog, Cat, Bird` - everything together

```python
# ═══════════════════════════════════════════════════════════════════════════════
# FULL DEMO - all OOP concepts in one realistic example
# ═══════════════════════════════════════════════════════════════════════════════

from abc import ABC, abstractmethod
from dataclasses import dataclass, field


class Animal(ABC):
    """Abstract base class - all animals must implement speak() and move()."""

    total_animals = 0       # Class variable

    def __init__(self, name: str, species: str, age: int):
        self.name      = name
        self.species   = species
        self._age      = age
        self.__id      = Animal.total_animals + 1
        Animal.total_animals += 1

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, value):
        if value < 0:
            raise ValueError("Age cannot be negative.")
        self._age = value

    @property
    def animal_id(self):
        return self.__id       # Read-only

    @abstractmethod
    def speak(self) -> str: ...

    @abstractmethod
    def move(self) -> str: ...

    def eat(self, food: str) -> str:
        return f"{self.name} enjoys {food}."

    def __repr__(self):
        return f"{self.__class__.__name__}(#{self.__id}, {self.name!r})"

    def __str__(self):
        return f"[{self.__class__.__name__} #{self.animal_id}] {self.name} | {self.species} | Age: {self.age}"

    def __eq__(self, other):
        return isinstance(other, Animal) and self.animal_id == other.animal_id

    def __lt__(self, other):
        return self.age < other.age


class Dog(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, "Canis lupus familiaris", age)
        self.breed  = breed
        self.tricks = []

    def speak(self): return f"{self.name}: Woof!"
    def move(self):  return f"{self.name} runs on 4 legs."

    def learn(self, trick):
        self.tricks.append(trick)
        return f"{self.name} learned '{trick}'!"

    def __str__(self):
        return super().__str__() + f" | Breed: {self.breed} | Tricks: {self.tricks}"


class Cat(Animal):
    def __init__(self, name, age, indoor=True):
        super().__init__(name, "Felis catus", age)
        self.indoor = indoor

    def speak(self): return f"{self.name}: Meow"
    def move(self):  return f"{self.name} prowls silently."
    def purr(self):  return f"{self.name} purrrrs... 💤"


class Bird(Animal):
    def __init__(self, name, age, can_fly=True):
        super().__init__(name, "Aves", age)
        self.can_fly = can_fly

    def speak(self): return f"{self.name}: Tweet tweet!"
    def move(self):
        return f"{self.name} flies through the sky!" if self.can_fly \
               else f"{self.name} waddles (can't fly)."


class Shelter:
    """Animal shelter - uses polymorphism to handle mixed animal types."""

    def __init__(self, name: str):
        self.name    = name
        self._animals = []

    def admit(self, animal: Animal) -> None:
        self._animals.append(animal)
        print(f"  ➕ Admitted: {repr(animal)}")

    def morning_roll_call(self) -> None:
        print(f"\n{'═'*50}")
        print(f"{self.name} - Morning Roll Call")
        print(f"{'─'*50}")
        for animal in sorted(self._animals):   # sorted by age (__lt__)
            print(f"  {animal.speak():<35} | {animal.move()}")
        print(f"{'═'*50}")
        print(f"  Total animals: {len(self._animals)}")

    def find_by_name(self, name: str):
        return next((a for a in self._animals if a.name == name), None)


# ── Run the shelter ───────────────────────────────────────────────────────────
print("=== ANIMAL SHELTER DEMO ===")
shelter = Shelter("Happy Paws Shelter")

d1 = Dog("Bruno",   3, "Labrador")
d2 = Dog("Max",     1, "Beagle")
c1 = Cat("Whiskers", 5)
c2 = Cat("Luna",     2, indoor=False)
b1 = Bird("Tweety", 1, can_fly=True)
b2 = Bird("Tux",    4, can_fly=False)

for animal in [d1, d2, c1, c2, b1, b2]:
    shelter.admit(animal)

d1.learn("sit")
d1.learn("shake hands")
d2.learn("roll over")

shelter.morning_roll_call()

# Find and interact
found = shelter.find_by_name("Whiskers")
if found:
    print(f"\nFound: {found}")
    print(found.purr())
    print(found.eat("tuna"))

# Polymorphism - one function works for all
print(f"\nTotal animals registered: {Animal.total_animals}")
```

# Day 3 - Complete OOP Summary

| Concept | Python feature | Remember |
|---------|---------------|----------|
| Class | `class Foo:` | Blueprint for objects |
| Constructor | `__init__(self, ...)` | Runs on `Foo()` |
| Instance var | `self.x = ...` | Unique per object |
| Class var | `Foo.x = ...` | Shared by all |
| Dunder methods | `__str__`, `__add__`, etc. | Python's operator hooks |
| `@property` | `@property` + `@x.setter` | Attribute with validation |
| `@classmethod` | `cls` as first param | Factory / alternative constructors |
| `@staticmethod` | No `self`/`cls` | Utility functions |
| Inheritance | `class Dog(Animal)` | Reuse parent code |
| `super()` | `super().__init__(...)` | Call parent method |
| MRO | `ClassName.__mro__` | Left-to-right method resolution |
| Polymorphism | Override `speak()` | Same call, different result |
| Duck typing | No type check | Just needs the method |
| ABC | `@abstractmethod` | Enforce interface |
| Encapsulation | `_protected`, `__private` | Convention-based access control |
| Dataclass | `@dataclass` | Auto `__init__`, `__repr__`, `__eq__` |
